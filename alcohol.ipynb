```python
import pandas as pd
import math
alcohol = pd.read_excel("data/alcohol.xlsx")
#get alcohol data from excel file
```

```python
for i in alcohol.columns[1:]:
    counter = 0
    for j in alcohol[i]:
        if math.isnan(j):
            alcohol[i][counter]=0
        counter+=1
```

```python
copy_of_alcohol = alcohol.copy()
for index, row in alcohol.iterrows():
    max_for_row = max(row[1:])
    for i in alcohol.columns[1:]:
        copy_of_alcohol[i][index] = row[i]/max_for_row
```
